# Ultimos detalles influenciados hacia la posterior construcción del DataBase en SQL

```python
data_film['actors'] = ''

def actors(movie): 
    act = []
    for i in data_old.index:
        if data_old.title[i] == movie: 
            name = data_old.first_name[i] + ' ' + data_old.last_name[i]
            act.append(name)
    return ', '.join(act)

for i in data_film.index:
    data_film['actors'][i] = actors(data_film.title[i])
```

```python
data_film['category'] = ''

def category(movie):
    categor = ''
    for i in data_old.index:
        if data_old.title[i] == movie:
            categor = data_cat[data_cat.category_id == data_old.category_id[i]].name.iloc[0]
            return categor

for i in data_film.index:
    data_film['category'][i] = category(data_film.title[i])

data_film.category.fillna('', inplace=True)
```
