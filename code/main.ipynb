# Querys

---

Procedemos a realizar algunas consultas que serán útiles en el futuro para nuestro videoclub con el fin de obtener distintos tipos de información alojada en nuestra base de datos.

```python
query = '''

-- Obtener todas las películas disponibles en el inventario:

SELECT film.title, film.description, film.rental_rate, inventory.inventory_id
FROM film
INNER JOIN inventory
ON film.film_id = inventory.film_id

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener la cantidad de películas rentadas por mes:

SELECT MONTH(rental.rental_date) AS mes, COUNT(*) AS cantidad
FROM rental

GROUP BY MONTH(rental.rental_date)

;
'''

datos = pd.read_sql(query, cursor)

datos
```

```python
query = '''

-- Obtener el top 10 de películas más rentadas:

SELECT film.title, COUNT(*) AS cantidad_rentas
FROM film
INNER JOIN inventory
ON film.film_id = inventory.film_id
INNER JOIN rental
ON inventory.inventory_id = rental.inventory_id

GROUP BY film.title
ORDER BY cantidad_rentas DESC
LIMIT 10

;
'''

datos = pd.read_sql(query, cursor)

datos
```

```python
query = '''

-- Obtener la lista de películas más populares ordenadas por su valoración:

SELECT film.title, film.rental_rate
FROM film

ORDER BY film.rental_rate DESC

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener la lista de películas disponibles para renta en la fecha actual:

SELECT film.title
FROM film
JOIN inventory
ON film.film_id = inventory.film_id
INNER JOIN rental
ON inventory.inventory_id = rental.inventory_id
WHERE inventory.inventory_id NOT IN (
  SELECT inventory_id FROM rental WHERE return_date IS NULL
)

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener todas las películas de una categoría determinada:

SELECT film.title, film.description, film.release_year, film.length, film.lead_actor_actress
FROM film
INNER JOIN category
ON film.category_id = category.category_id
WHERE category.name = 'Drama'

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener el total de alquileres realizados por cada cliente:

SELECT customer_id, COUNT(*) AS total_alquileres
FROM rental

GROUP BY customer_id
ORDER BY total_alquileres DESC

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener las películas más largas (con mayor duración) de cada categoría:

SELECT categoria, pelicula, duracion
FROM (
    SELECT
        category.name AS categoria, film.title AS pelicula, film.length AS duracion,
        ROW_NUMBER() OVER (PARTITION BY film.category_id ORDER BY film.length DESC) AS row_num
    FROM film
    JOIN category 
    ON film.category_id = category.category_id
) AS ranked_films

WHERE row_num = 1;

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener el promedio de duración de películas para cada categoria:

SELECT category.name AS categoria, AVG(film.length) AS duracion_promedio
FROM film
JOIN category
ON film.category_id = category.category_id

GROUP BY category.name

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```

```python
query = '''

-- Obtener la cantidad de películas en cada categoría:

SELECT category.name AS categoria, COUNT(*) AS cantidad_peliculas
FROM film
JOIN category 
ON film.category_id = category.category_id

GROUP BY category.name

;
'''

datos = pd.read_sql(query, cursor)

datos.head()
```
