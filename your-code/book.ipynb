```python
data_film['actors'] = '' # creamos una columna vacia, llamada 'actors'

def actors(movie): 
    act = []
    for i in data_old.index:  # creamos una funcion que saque los actores y actrices de esa pelicula
        if data_old.title[i] == movie: 
            name = data_old.first_name[i] + ' ' + data_old.last_name[i]
            act.append(name)
    return ', '.join(act)

for i in data_film.index:  # introducimos los nombres por cada fila
    data_film['actors'][i] = actors(data_film.title[i])
```

```python
data_film.head()
```

```python
data_film['category'] = '' # creamos una columna vacia, llamada 'category'

def category(movie):  # creamos una funcion que saque la categoría de esa pelicula
    categor = ''
    for i in data_old.index:
        if data_old.title[i] == movie:
            categor = data_cat[data_cat.category_id == data_old.category_id[i]].name.iloc[0]
            return categor

for i in data_film.index:  # introducimos las categorias por cada fila
    data_film['category'][i] = category(data_film.title[i])

data_film.category.fillna('', inplace=True) # y rellenamos los valores 'None' de category con '' (nada)
```
